```python
df = pd.read_csv('DataFiles/Match.csv')
df1 = pd.read_csv('DataFiles/Ball_by_Ball.csv')
df2 = pd.read_csv('DataFiles/Player.csv')
df3 = pd.read_csv('DataFiles/Player_match.csv')
df4 = pd.read_csv('DataFiles/Team.csv')
tid=df4[df4['Team_Name'] == "Mumbai Indians"]['Team_Id']
mat_id=df1['Match_Date'] == "4/23/2008"
bat=df1['Team_Batting'] == str(int(tid))
d=df1[mat_id & bat][['Striker','Runs_Scored']]
x=d.groupby(['Striker']).Runs_Scored.sum()
k=x.to_dict()
li=list(x.keys())
matchid=df[df['match_date'] == "4/23/2008"]['match_id']
team=df3['Player_team'] == 'Mumbai Indians'
y=df3['Match_Id'] == int(matchid)
m=df3[y & team]['Player_Id']
n=df3[y & team]['Player_Name']
li1=list(m)
li2=list(n)
for i in range(0,len(li1)):
    if li1[i] not in li:
        a=li1[i]
        k[a]=0

li1=list(k.values())
def bar_graph(u,w):
        x=1/2
        plt.bar(u,w,x,color=('Purple','darkblue','skyblue','pink','violet'))
        plt.rcParams['figure.figsize']=[15,5]
        plt.xlabel('PlayerList')
        plt.ylabel('Runs')
        plt.title('Bar graph for all the players in the Mumbai Indians team')
        plt.show
res=bar_graph(li2,li1)
```
