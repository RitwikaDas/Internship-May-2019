#### Problem 1:
#### Problem statement:
Draw a pie plot based on number of wins for each team


```python
def Matchwin(filename):
    df = pd.read_csv(filename)
    x = df['match_winner'].value_counts()
    d = dict(x)
    li1=list(d.keys())
    li2=list(d.values())
    labels = li1
    sizes = li2
    fig1, ax1 = plt.subplots()
    ax1.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
    ax1.axis('equal')
    plt.show()
Matchwin('DataFiles/Match.csv')
```

```python
class Matchwin():
    def abc(self,filename):
        df = pd.read_csv(filename)
        x =df['match_winner'].value_counts()
        d = dict(x)
        self.li1=list(d.keys())
        self.li2=list(d.values())
        return self.pie()
    def pie(self):
        labels =self.li1
        sizes = self.li2
        fig1, ax1 = plt.subplots()
        ax1.pie(sizes,labels=labels,autopct='%1.1f%%',shadow=True,startangle=90)
        #fig1 = plt.gcf()
        #fig1.set_sizeinches[5,5]
        ax1.axis('equal')
        plt.show()
obj = Matchwin()
obj.abc('DataFiles/Match.csv')
```
